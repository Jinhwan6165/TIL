```python
import pandas as pd
```

```python
A = pd.read_excel('./data/exam.xlsx')
A
```

```python
B = pd.read_excel('./data/exam_extra.xlsx')
B
```

```python
pd.merge(A, B, how = 'left', left_on='번호', right_on='번호')  
# how = 어느 테이블 기준으로 합칠건지..
# how = left, right, inner, outer
# A, B or B,A 처럼 순서에 따라 원하는 것을 기준으로 merger 가능
```

```python
pd.merge(A, B, how = 'right', left_on='번호', right_on='번호')  
```

```python
pd.merge(A, B, how = 'inner', left_on='번호', right_on='번호')  
```

```python
df = pd.merge(A, B, how = 'outer', left_on='번호', right_on='번호')  
df = pd.merge(A, B, how = 'outer', left_index='번호', right_on='번호')  

#df = pd.merge(A, B, how = 'outer', left_on='컬럼명', right_on='번호')  
#df = pd.merge(A, B, how = 'outer', left_indeex='True', right_on='True')  
# 번호가 인덱스에 있다면.... 
# left_on과 right_on이 같은 컬럼이면 on='번호'로 할 수 있다.
df
```
