## Scatter Matrix Plot

* 튜토리얼 영상링크 : https://youtu.be/PtSofOihtZs # Scatter Matrix Plot
* https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html#plotting-tools

```python
import pandas as pd
import numpy as np

import matplotlib as mpl

%matplotlib inline

#한글폰트 사용시 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
from pandas.plotting import scatter_matrix
df = pd.DataFrame(np.random.randn(1000, 4), columns=['a', 'b', 'c', 'd'])
df.head()
```

```python
# kernel density estimate 커널 밀도 함수 의미
scatter_matrix(df, alpha=0.2, figsize=(6,6),diagonal='kde')
```
