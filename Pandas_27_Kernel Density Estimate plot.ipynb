### 정규분포
* 확률론과 통계학에서, 정규 분포(normal distribution) 또는 가우시안 분포(Gaussian distribution)는 연속 확률 분포의 하나이다. 정규분포는 수집된 자료의 분포를 근사하는 데에 자주 사용되며, 이것은 중심극한정리에 의하여 독립적인 확률변수들의 평균은 정규분포에 가까워지는 성질이 있기 때문이다. - 출처 : [정규 분포 - 위키백과, 우리 모두의 백과사전](https://ko.wikipedia.org/wiki/%EC%A0%95%EA%B7%9C_%EB%B6%84%ED%8F%AC)

```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

#한글폰트 사용시 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
ser = pd.Series(np.random.randn(1000))
```

```python
ser.plot.hist()
```

```python
# 곡선으로 표현, density는 y축에 히스토그램처럼 frequency가 아니라 density가 있다.
ser.plot.density()
```

```python
ser.plot.kde()
```
