#### 성별 비중 계산하기


```python
name_df['total'] = name_df['M']+ name_df['F']
name_df.head()
```

```python
name_df['F_ratio'] = name_df['F']/ name_df['total']
name_df['M_ratio'] = name_df['M']/ name_df['total']
name_df.head ()
```

```python
cond = (name_df['total']>10000)\
    & (name_df['M_ratio']>0.4)\
    & (name_df['M_ratio']<0.6)
name_df[cond].sort_values('total', ascending=False)
```
