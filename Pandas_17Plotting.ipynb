파이썬 판다스로 Series와 DataFrame 데이터 시각화 하기 소개

## plot
  ### kind : str
    The kind of plot to produce:

    - 'line' : line plot (default)
    - 'bar' : vertical bar plot
    - 'barh' : horizontal bar plot
    - 'hist' : histogram
    - 'box' : boxplot
    - 'kde' : Kernel Density Estimation plot
    - 'density' : same as 'kde'
    - 'area' : area plot
    - 'pie' : pie plot
    - 'scatter' : scatter plot
    - 'hexbin' : hexbin plot.

### Basic plotting

```python
ts = pd.Series(np.random.randn(1000),
              index=pd.date_range('1/1/2000', periods=1000))
ts
```

```python
ts.plot()
```

```python
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000,4),
              index=ts.index, columns=list('ABCD'))
```

```python
#df
df.shape
```

```python
df.plot()
```

```python
df = df.cumsum()
```

```python
df.plot()
```

```python
df3 = pd.DataFrame(np.random.randn(1000,2), columns=['B','C']).cumsum()
```

```python
df3.head()
```

```python
df3['A'] = pd.Series(list(range(len(df))))   # A컬럼 추가 생성
```

```python
df3.head()
```

```python
df3.plot(x='A',y='B')
```
