### 일부 컬럼을 기준으로 데이터 가져오기 Subset Variables(Columns)

```python
import seaborn as sns
```

```python
df = sns.load_dataset("iris")

df.head()
```

```python
# 특정 컬럼을 가져올 때 []에 담아서 리스트 형태로 가져온다.
columns = ['sepal_width','sepal_length','species']
df[columns].head()
```

```python
# df['sepal_width'] or df.sepal_width
df['sepal_width'] or df.sepal_width

# 한글 또는 컬럼명에 특수문자로 가져올 때 오류가 나는 경우 []로 다시 감싸준다.
# df['sepal_width']
```

```python
df.filter(regex='^(?!species$).*') # regex : 정규 표현식으로 조건을 써주면 해당하는 컬럼 추출

# '\.' : 조건에 해당하는 내용 추출
# 'Length$' : $ 앞 단어로 끝나는 컬럼 추출
# '^Sepal' : 특정 문자로 시작하는 것 추출
# '^x[1-5]$' : x로시작해서 1-5로 끝나는 것 추출 ex('es$'')
# '^(?!Species$).*' : 특정문자를 제외하고 나머지 
```

```python
df.loc[2:5,'sepal_width':'petal_width'] #범위 지정하여 가져오기 앞=행/뒤=열
# 해당 인덱스 포함하여 가져옴
```

```python
# 인덱스 순서대로 가져오기 ( 해당 인덱스 전까지 가져옴)
df.iloc[:-5,[1,2,4]]
```

```python
df.loc[df['sepal_length'] > 5, ['sepal_length','sepal_width']]
#                       행,         열
# 인덱스 포함하여 가져온다.
```

```python
df.iloc[-5:,[1,3]] #iloc : 인덱스 번호 순으로 가져옴, 해당 인덱스 전까지 가져욤
# : = 전체데이터를 의미한다.
```

```python
df.loc[df['sepal_length']>5, ['sepal_length','sepal_width']]
```
